```python
light = pd.read_csv('Datasets/light.csv')
co2 = pd.read_csv('Datasets/Co2.csv')
dateTime = light[['Date', 'Time']]

def To24(df):
    return pd.to_datetime(df, format='%I:%M %p').dt.strftime('%H:%M:%S');

def ToDateTime(Date, Time):
    return pd.to_datetime(Date + ' ' + Time)


def NormalizeData(df, Columns):
    df[Columns] = (df[Columns]-df[Columns].mean())/df[Columns].std()
        
#Change time to 24 hours



light['Time']=To24(light['Time'])
light['DateTime'] = ToDateTime(light['Date'], light['Time'])

co2['Time']=To24(co2['Time'])
co2['DateTime'] = ToDateTime(co2['Date'], co2['Time'])

NormalizeData(light, ['Value'])
NormalizeData(co2, ['Inst','TWA'])
#NormalizeData(co2, ['TWA'])
#DateTime = pd.to_datetime(light['Date'] + ' ' + light['Time'])
#print(co2.to_string()) 
from matplotlib import pyplot as plt
#plt.plot(light['DateTime'], light['Value'],label = "light")
plt.plot(co2['DateTime'], co2['Inst'],label = "co2")
plt.plot(co2['DateTime'], co2['TWA'],label = "Twa")
plt.legend()
plt.show()
```
